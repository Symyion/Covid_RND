```python
import pandas as pd
import seaborn as sns
import numpy as np
import plotly.express as px
import py_vollib.black_scholes.implied_volatility as pv
```

```python
project_path = '/Users/symyion/PycharmProjects/master_thesis/'
```

```python
raw_data = pd.read_csv(project_path + 'data/input/sp500_data.csv')
raw_data
```

```python
s = 4226.25
```

```python
raw_data = raw_data.assign(moneyness_log = np.abs(np.log(raw_data["strike"] / s)))

df = raw_data.loc[(raw_data['volume'] > 0) &
                  (raw_data['bid'] > 0.5) &
                  (raw_data["moneyness_log"] <= np.log(1.4))]
df
```

```python
fig = px.scatter(df,
                  x = "strike",
                  y="bid",
                  color = "put_call")
fig.show()
```

```python
def implied_vol(row):
    S = s
    price = row['bid']
    K = row['strike']
    t = 11/252
    r = 0.001
    q = 0.0136
    if row['put_call'] == 'put':
        flag = 'p'
    else:
        flag = 'c'
    IV = pv.implied_volatility(price, S, K, t, r,flag, return_as='series')
    return IV
```

```python
df['IV'] = df.apply(implied_vol,axis=1)
df.dropna(inplace=True)
```
